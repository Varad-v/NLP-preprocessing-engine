## 📌 Conceptual Understanding

### 1. Difference between "Love" and "love"
"Love" and "love" are treated as different tokens due to case sensitivity in NLP. This can lead to duplication and inconsistency. Lowercasing solves this issue.

### 2. What happens if stopwords are not removed?
Stopwords increase noise, reduce model performance, and add unnecessary features.

### 3. When removing stopwords is harmful?
- Sentiment Analysis → "not happy" becomes "happy"
- Question Answering → "what is your name" loses meaning

### 4. Stemming vs Lemmatization
Stemming cuts words using rules (running → runn), while lemmatization returns meaningful base forms (running → run).

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    """
    Cleans and preprocesses text into tokens.
    Handles noise such as URLs, numbers, emojis, and repeated characters.
    """

    if not isinstance(text, str) or text.strip() == "":
        return []

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove emojis & special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens except 'no' and 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return tokens
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",
    "",  # Empty case
    "😂😂😂",  # Only emojis
    "1234567890"  # Only numbers
]
```

```python
print("🔍 STRESS TEST RESULTS\n")

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sentence)
    print("-" * 60)
```

```python
def analyze_tokens(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)
```

```python
print("📊 TOKEN ANALYTICS\n")

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    total, unique, avg_len = analyze_tokens(tokens)

    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Average Token Length: {avg_len}")
    print("-" * 60)
```

```python
all_tokens = []

for sentence in test_sentences:
    all_tokens.extend(preprocess_text(sentence))

counter = Counter(all_tokens)

print("🔥 Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\n❄️ Least 5 Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("📦 PIPELINE OUTPUT\n")
print("Tokens:", result["tokens"][:20])
print("\nClean Sentences:", result["clean_sentences"])
```
